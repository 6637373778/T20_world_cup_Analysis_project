# Win percentage of all teams in ICC T20 World Cup

```python
all_teams = pd.concat([matches['team1'], matches['team2']])    # represent the teams playing in each match.
total_matches=all_teams.value_counts()
matches_won=matches['winner'].value_counts()
win_percentage=(matches_won/total_matches)*100
print(win_percentage)
```

1. The pd.concat()- function combines these two series into a single series all_teams that contains all the teams that participated in the matches, including both teams in each match (both team1 and team2).
2. total_matches = all_teams.value_counts()- This calculates the total number of matches each team has played by counting how many times each team appears in the all_teams series.
3. matches_won = matches['winner'].value_counts()- This line calculates how many matches each team has won by counting the number of occurrences of each team in the winner column of the matches DataFrame.
4. win_percentage = (matches_won / total_matches) * 100 - The formula (matches_won / total_matches) * 100 divides the number of matches a team won by the total number of matches they played, and then multiplies the result by 100 to convert it into a percentage.

```python
plt.xlabel("Country")
plt.ylabel("Win percentage")
plt.title("Win percentage of teams")
win_percentage.plot(kind="bar")
```

# Top Wicket takers

```python
wicket_takers = deliveries[deliveries['wicket_type'].notnull()]
wicket_takers_count = wicket_takers.groupby('bowler').size().reset_index(name='wickets')
wicket_takers_sorted = wicket_takers_count.sort_values(by='wickets', ascending=False).reset_index(drop=True)

wicket_takers_sorted
```

1. wicket_takers = deliveries[deliveries['wicket_type'].notnull()]- This line filters the deliveries DataFrame to include only the rows where a wicket was taken.
2. wicket_takers_count = wicket_takers.groupby('bowler').size().reset_index(name='wickets') - groupby('bowler'): This groups the wicket_takers DataFrame by the bowler column, so that all deliveries by each bowler are grouped together.
3. wicket_takers_sorted = wicket_takers_count.sort_values(by='wickets', ascending=False).reset_index(drop=True) - This sorts the wicket_takers_count DataFrame by the wickets column in descending order, so the bowler with the most wickets comes first.
4. wicket_takers_sorted - This will display the final DataFrame containing the bowlers and their wicket counts, sorted in descending order of the number of wickets taken.
